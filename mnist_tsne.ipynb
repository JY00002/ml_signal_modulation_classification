This notebook shows a different example on how to use tsne_utils 
using the model trained in mnist_example to classify handwritten digits from the
mnist dataset.

```python
from keras.models import load_model
from tsne_utils import *
```

```python
data = np.load('mnist.npz')
x_test, y_test = data['x_test'], data['y_test']
```

```python
print(x_test.shape, y_test.shape)
```

```python
# get 100 samples for each label

t_data = []
t_labels = []

num_points = 100

for i in range(10):
    idx = np.where(y_test == i)[0][:num_points]
    t_data.append(x_test[idx])
    t_labels.append(np.char.mod('%d', y_test[idx]))
    
t_data = np.concatenate(t_data)
t_labels = np.concatenate(t_labels)
    
print(t_data.shape)
print(t_labels.shape)
```

```python
# load the model
model = load_model('mnist_model.h5')
```

```python
BOOKEH_COLORS = {
    '0': 'aqua', 
    '1': 'aquamarine',  
    '2': 'bisque',  
    '3': 'black', 
    '4': 'blue', 
    '5':'blueviolet', 
    '6': 'brown',
    '7': 'burlywood', 
    '8': 'cadetblue', 
    '9': 'chartreuse', 
    '10': 'chocolate'
}


BOOKEH_SHAPES = {
    '0': 1, 
    '1': 1,  
    '2': 1,  
    '3': 1, 
    '4': 1, 
    '5':1, 
    '6': 1,
    '7': 1, 
    '8': 1, 
    '9': 1, 
    '10': 1
}

classes = np.char.mod('%d', np.arange(24))
```

```python
tsne_plot(model, t_data, t_labels, classes, BOOKEH_COLORS, BOOKEH_SHAPES,layer_index=-2, legend = True, show_preds=True)
```
